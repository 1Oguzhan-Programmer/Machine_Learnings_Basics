# scikit-learn kütüphanesi decision tree'lerin düzgün çalışması için herşeyin rakamsal olmasını bekliyor bu nedenle veri setimizdeki tüm Y ve N değerlerini 0 ve 1 olarak düzeltiyoruz. Aynı sebeple eğitim seviyesini de Bs:0 Ms:1 ve PhD:2 olarak güncelliyoruz. map() kullanarak boş hücreler veya geçersiz değer girilen hücreler NaN ile doldurulacaktır. Buna şuandaki veri setimizde ihtiyacımız yok ama sizin ileride yoğun veri ile çalıştığınız zaman ihtiyacınız olacaktır. 



# scikit-learn library expects everything to be numeric for decision trees to work properly, so we fix all Y and N values ​​in our dataset to 0 and 1. For the same reason, we update the education level to Bs:0 Ms:1 and PhD:2. Empty cells or cells with invalid values ​​will be filled with NaN using map(). We don't need this in our current dataset, but you will need it in the future when you work with large amounts of data.

```python
duzeltme_mapping = {"Y": 1, "N": 0}

df["IseAlindi"] = df["IseAlindi"].map(duzeltme_mapping)
df["Suancalisiyor?"] = df["SuanCalisiyor?"].map(duzeltme_mapping)
df["Top10 Universite?"] = df["Top10 Universite?"].map(duzeltme_mapping)
df["StajbizdeYaptimi?"] = df["StajBizdeYaptimi?"].map(duzeltme_mapping)
duzeltme_mapping_egitim = {"BS":0, "MS":1, "PhD":2}
df["Egitim Seviyesi"] = df["Egitim Seviyesi"].map(duzeltme_mapping_egitim)
df.head()
```
