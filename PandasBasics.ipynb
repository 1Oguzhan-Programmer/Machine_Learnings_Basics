                                             # VERİ FİLTRELEME (DATA FILTERING)

# Baba filminin IMBD reyting'ini bulalım. Eğer filmin dataframe'deki sırasını biliyorsak basitçe dataframe'in IMBD_Rating sütununun 2. elemanını (ilk index 0 olduğu için 1 diye yazıyoruz, listelerde ilk eleman 0'dan başlardı hatırlarsınız.)

# Let's find the IMBD rating of The Godfather movie. If we know the order of the movie in the dataframe, we simply add the 2nd element of the IMBD_Rating column of the dataframe (we write it as 1 because the first index is 0, you may remember that the first element in the lists starts from 0)

```python
df["IMDB_Rating"][1]
```

# Ancak eğer  baba filminin veri setini görme şansımız yok yani büyük veri seti var içinde bir yerlerde o zaman kodla bulabiliriz. 
# However, if we do not have a chance to see the data set of the movie The Godfather, there is a large data set somewhere in it, then we can find it with code.

```python
df.loc[df["Series_Title"]=="the Godfather"]
```

```python
# loc fonksiyonu konumu bildirir.
# The loc function reports the location.






```

```python
# Bunu nasıl yaptığımızı anlatalım. Öncelikle df.loc["Series_Title"]=="The Godfather"] ile aslında baba filminin olduğu satırı buluyoruz.
# Let's explain how we did this. First of all, with df.loc["Series_Title"]=="The Godfather"] we find the line where the Godfather movie is actually located.
```

```python
df.loc[df["Series_Title"]=="The Godfather"]
```

# Ama bizim istediğimiz sadece IMDB_Rating değeri yani tüm satırı istemiyoruz, bu satırın içinden "IMDB_Rating" sütununu seçiyoruz Bize df.loc ile dönen bir değişken var o değişken aslında bir dataframe'dir.

# But what we want is only the IMDB_Rating value, that is, we do not want the entire row, we select the "IMDB_Rating" column from this row. There is a variable that returns us with df.loc. That variable is actually a dataframe.

```python
type(df.loc[df["Series_Title"]=="The Godfather"])
```

```python
# O zaman bir dataframe'in basitçe sütununu seçmek için köşeli parantez kullanıyor ve seçiyoruz.
# Then we use square brackets to simply select a column of a dataframe and select.
```

```python
df.loc[df["Series_Title"]=="The Godfather"]["IMDB_Rating"]
```

```python
# Son olarak da  IMDB_Rating değeri 8'in üstünde olan ve No_of_Votes (Oy Sayısı) 100000'den  (100 bin) çok olna filmleri görelim.
# Finally, let's see the movies with an IMDB_Rating value of over 8 and a No_of_Votes of more than 100,000 (100 thousand).
```

```python
df.loc[(df["IMDB_Rating"]>=8) & (df["No_of_Votes"]>=100000)]      # & ve işaretidir.
```

```python
# Şimdi bir de IMDB_Rating değeri ('in üstünde olan Gross değeri (Hasılat) 30000000'dan (30 milyon) fazla olan filmleri görelim.  
# Now let's look at the movies with an IMDB_Rating value above ( and a Gross Value (Revenue) of more than 30000000 (30 million).
```

```python
df.loc[(df["IMDB_Rating"]>=8) & (df["Gross"]>="30000000")]        # Gross'u string olarak girdiği için string şeklinde baktık.
```
