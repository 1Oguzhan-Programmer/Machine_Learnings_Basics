```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Temel Çizim İşlemleri (Basic Drawing Procedures)



# Plot y versus x as lines and/or markers. plot(*args[,scalex, scaley, data])  (X eksenine göre y'yi çizicez)
```

```python
# plt, matplotlib'in kısaltılmş hali 
# plt, short for matplotlib
```

```python
plt.plot([1,2,3,4], [1,4,9,16])   # Birincisi X ekseni, İkincisi Y ekseni (# First is X axis, Second is Y axis)
```

```python
x = [1,2,3,4]
y = [1,4,9,16]


plt.plot(x,y)
plt.show()
```

```python
# Grafiğe, x eksenine ve y eksenine başlık ekleyeceğiz. 
# We will add titles to the chart, x-axis and y-axis.
```

```python
plt.title("İlk grafiğimiz")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.plot(x,y)
plt.show()
```

```python
# Normalde çentikler otomatik olarak yerleştirilir.
# Normally notches are placed automatically.

# Kendiniz de çentiklerin yerini tanımlayabilirsiniz
# You can also define the location of the notches yourself

# Küsüratları da gösterebilir.
# It can also show fractions.
```

```python
plt.xticks([1,2,3,4])
plt.yticks([1,4,9,16])
plt.title("İlk grafiğmiz")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show(x,y)
plt.show()
```

                                                                Lejant Koyma

# Çizdiğiniz grafiğe lejant koyma. Lejant, haritada kullanılan özel işaretlerin ne anlama geldiğini gösteren bölümdür. 
# Putting a legend on the graph you draw. The legend is the section that shows what the special signs used on the map mean.

```python
plt.plot(x,y, label="x^2", color="green")
plt.xticks([1,2,3,4,5])
plt.yticks([1,4,9,16,25])
plt.title("İlk grafiğimiz")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend()
plt.show()
```

```python
# Aynı grafikte birden fazla çizim yapmak                        (axis = Eksen anlamında)
# Drawing multiple plots on the same chart
```

```python
plt.plot(x,y, label="x^2", color="green", linewidth = 2, linestyle = "--", marker=".")
plt.xticks([0,1,2,3,4,5])
plt.yticks([0,1,4,9,16,25])
plt.title("İlk grafiğimiz")
plt.xlabel("x axis")
plt.ylabel("y axis")



# 2. çizgimizi çiziyoruz.
# We draw our 2nd line.

x2 = np.arange(0,5,0.5)   #np.arange NumPy kütüphanesinde belirli bir aralıktaki sayıları içeren bir dizi oluşturmayı sağlayan bir fonskiyondur.
plt.plot(x2,x2*2, color="red", linewidth = 2, marker=".", label="2*x")

plt.legend()
plt.show()
```

```python
# Grafiğinizi bilgisayarınıza kaydetme 
# Save your chart to your computer
# savefig komutu ile olur isim veririz ve kaydederiz.
```

                                                             # Barchart
 # Bir barchart, kategorik verilerin frekansını veya dağılımını görselleştirmek için kullanılan bir grafik türüdür. Bu grafik türü, her bir kategoriye ait değerlerin çubuklarla temsil edildiği ve bu çubukların yüksekliğinin kategorinin değeriyle orantılı olduğu bir grafiktir.   

 # A barchart is a type of chart used to visualize the frequency or distribution of categorical data. This type of chart is a chart where the values ​​of each category are represented by bars and the height of these bars is proportional to the value of the category.

```python
x = ["Ankara", "İstanbul", "İzmir"]
y = [120, 178,87]

plt.bar(x,y) 
plt.show()

# plt.bar(x,y) ifadesi, Matplotlib kütüphanesini kullanarak bir barchart oluşturmak için kullanılan bir fonksiyondur.
# Bu fonskiyon, x ve y değerlerini alır ve bu değerleri kullanarak çubukları çizmeyi sağlar. 
# plt.bar(x,y) expression is a function used to create a barchart using the Matplotlib library.
# This function takes x and y values ​​and allows drawing bars using these values.
```

# Eğer çubuklardan bir tanesini diğerlerinden farklılaştırmak için içini boyabilirsiniz.
# If you want to differentiate one of the sticks from the others, you can paint the inside.

```python
x = ["Ankara", "İstanbul", "İzmir"]
y = [120, 178, 87]

cubuklar = plt.bar(x,y)
cubuklar[0].set_hatch("/")    #Grafiğin en solundaki index 0'dır. Diğeri 1 sonrakide 2 dir
cubuklar[1].set_hatch(".")
cubuklar[2].set_hatch("//")
plt.show()
```

```python
gas = pd.read_csv("petrolfiyatlari.csv")           # (Bu csv dosyasının oluşturduğunuz dosyaya atmalısınız.)
gas                                                # (You should add this csv file to the file you created.)
```

```python
# bu veri setini çizdiriceğiz. x ekseni year, y ekseni usa olacak.title fonksiyonu ile cizicez. 
# 'b-' çizgi kelinde mavi çizgi cizdiricez anlamında. 

# We will plot this data set. The x axis will be year and the y axis will be usa. We will draw it with the title function. 
# 'b-' means we will draw a blue line on the dash.
```

```python
plt.title("petrol fiyatlari")
plt.plot(gas["Year"], gas["USA"], "b-", label="USA")

plt.xlabel("Yıl")
plt.ylabel("Dolar")

plt.legend()
plt.show()
```

```python
# Diğer ülkeleride görelim.
# Let's see other countries too.
```

```python
plt.title("petrol Fiyatlari")

plt.plot(gas.Year, gas.USA, "b-", label="USA")
plt.plot(gas.Year, gas["Canada"], "r-", label="Kanada")
plt.plot(gas.Year, gas["South Korea"], "g.-", label="Güney Kore")
plt.plot(gas.Year, gas["France"], "y.-", label="Fransa")

plt.xlabel("Yıl")
plt.ylabel("Dolar")

plt.legend()

plt.show()
```

```python
# gas istenilen sütunu verir. (Year ve USA gibi)   (# gas returns the desired column. (like Year and USA)
```
