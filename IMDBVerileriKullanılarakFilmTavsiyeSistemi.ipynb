                                           # Recommendation System (Öneri Sistemi)
# Bu sistemin temelinde iki tane algoritma yatıyor.  
# There are two algorithms underlying this system.

# 1) User-Based Collaborative Filtering (Kullanıcı Tabanlı İşbirlikçi Filtreleme)
# 2) Item-Based Collaborative Filtering (Öğe Tabanlı İşbirlikçi Filtreleme)




                                         1) User-Based collaborative Filtering

# Kullanıcılar arasınadaki benzerlik skorlarını bulmaya çalışıyor.

# Her bir kullanıcının satın aldığı veya izlediği/dinlediği materyali kullanıcı bazlı olarak tutar ve bir matrise dönüştürür.

# Bir kullacıya(X) benzer başka kullanıcılar bulur.

# X kullanıcısının henüz izlemediği ancak benzer kullanıcıların çoğunun izlediği bir film varsa bunu X kullanısına önerir.



# Trying to find similarity scores between users.

# Keeps the material purchased or watched/listened to by each user on a per-user basis and converts it into a matrix.

# Finds other users similar to a user(X).

# If there is a movie that user X has not watched yet but most of similar users have watched, it recommends it to user X.







# Kullanıcıları baz alır. 
# Kullanıcılar arasındaki benzerlikleri bulur buna göre önerme yapar.

# It is based on users.
# It finds similarities between users and makes suggestions accordingly.



# Dezavantajları (Disadvantages)

# Kullanıcı bazlı olduğu için maliyetlidir. İnsan sayısı film sayısından her zaman fazladır.
# İnsanların zevkleri değişebilir bugün sevdiğiniz filmi yarın sevmeyebilirsiniz. 


# It is costly because it is user-based. The number of people is always greater than the number of movies. 
# People's tastes may change, the movie you like today may not be liked tomorrow.



                                         2) Item-Based Colaborative filtering

# User-based collaborative fileting'deki sıkıntıları aşmak için Item-based metod geliştirilmiş ve kullanılmaktadır.

# Bizim bir sonraki ders python'da yapacağımız örnek de Item-based collaborative filtering metodudur. 


# To overcome the problems in user-based collaborative filing, the Item-based method has been developed and used.

# The example we will do in our next lesson in Python is the Item-based collaborative filtering method. 




# Kullanıcılar yerine Item'ları baz alır. Kullanıcılar arasındaki ilişkiyle ilgilenmez, Item'lar arasındaki ilişki ve benzerlikle ilgilenir. Yani birbirine benzeyen Item'ları bulur.

# Itemlar arasındaki benzerlik skorlarını bulmaya çalışır.

# Bizim film örneğimizde de eğer filmleri seven ortak kişiler varsa demek ki bu filmler birbirine benzer diye düşünür ve bu filmleri biribirine benzerlik skorlarını yüksek olarak belirler.

# Daha sonra siz bir film izlediğinizde o film için benzerlik skoru en yüksek olan filmleri size izlemeniz için önerir.






# It is based on Items instead of Users. It is not interested in the relationship between users, it is interested in the relationship and similarity between Items. In other words, it finds Items that are similar to each other.

# It tries to find the similarity scores between Items.

# In our movie example, if there are people who like movies in common, it means that these movies are similar to each other and determines these movies with high similarity scores.

# Later, when you watch a movie, it recommends the movies with the highest similarity scores for that movie.





# Item = öge (anlamında)




# Daha çok kullanılan sistem algoritması Item-Based Collaborative Filtering sistemi kullanılır.



                                   

```python
                                             # Recommendation Systems Project

import numpy as np
import pandas as pd
```

```python
column_names = ["user_id", "item_id", "rating", "timestamp"]
df = pd.read_csv("users.data", sep="\t", names=column_names)
df.head()


# sep="\t" ifadesi, print() fonksiyonunda birden fazla argüman yazdırıldığında bunlar arasına tab boşluğu koyar.
# The statement sep="\t" puts a tab space between multiple arguments printed in the print() function.

```

```python
# Kaç kayıt varmış görelim.
# Let's see how many records there are.

len(df)
```

```python
### Şimdi diğer dosyamızı yükleyelim. (### Now let's load our other file.)
```

```python
movie_titles = pd.read_csv("movie_id_titles.csv")
movie_titles.head()
```

```python
# Kaç kayıt varmış görelim:

len(movie_titles)
```

```python
df = pd.merge(df, movie_titles, on="item_id")
df.head()     # merge = Birleştir (anlamında)


# iki dataframe'i birleştirip tek dataframe üzerinde gideceğiz.
# We will merge two dataframes and move on to a single dataframe.

# Birleşirmedeki amacı buydu.
# That was the purpose of the merger.
```

```python
                                 # Recommendation Sistemimizi Kuruyoruz (We are Establishing Our Recommendation System)

# Öncelikle Excel'deki pivot tablo benzeri bir yapı kuruyoruz.
# Bu yapıya göre her satır bir kullanıcı olacak şekilde (yani dataframe'imizin index'i user_id olacak)
# Sütunlarda film isimleri (yani title sütunu) olacak
# Tablo içinde de rating değerleri olacak şekilde bir dataframe oluşturuyoruz.



# First, we create a structure similar to the pivot table in Excel.
# According to this structure, each row will be a user (i.e. the index of our dataframe will be user_id)
# The columns will contain movie names (i.e. the title column)
# We create a dataframe with rating values ​​in the table.


moviemat = df.pivot_table(index= "user_id", columns= "title", values="rating")
moviemat.head()   # Bir önceki user_id tablsounda excel yapılan pivot tablo oluşturdu.
```

```python
type(moviemat)     # moviemat aslında bir dataframe veri türü oldu.
```

```python
# Şimdi Star Wars (1977) filminin user ratinglerine bakalım.
```

```python
starwars_user_ratings = moviemat["Star Wars (1977)"]
starwars_user_ratings.head()
```

```python
# corwith() metodunu kullanarak Star Wars filmi ile korelasyonları hesaplayalım: Amaç= Starwars filmine benzer film önerileri yapmak.

# Let's calculate correlations with the Star Wars movie using the corwith() method: Goal = To make movie recommendations similar to the Starwars movie.
```

```python
similar_to_starwars = moviemat.corrwith(starwars_user_ratings)
similar_to_starwars
```

```python
# corwith() metodu pandas metodudur ve korelasyon hesaplar.
# The corwith() method is a pandas method and calculates correlation.
```

```python
type(similar_to_starwars)
```

# Bazı kayıtlarda boşluklar olduğu için hata veriyor similar_to_starwars bir seri yani dizi yani string biz bunu corr_starwars isimli bir dataframe'e dönüştürelim ve NaN kayıtlarını temizleyip bakalım.

# It gives an error because there are spaces in some records. similar_to_starwars is a series, that is, a string. Let's convert it to a dataframe named corr_starwars and clean the NaN records and see.







```python
corr_starwars = pd.DataFrame(similar_to_starwars, columns=["Correlation"])
corr_starwars.dropna(inplace=True)
corr_starwars.head()
```

```python
# Elde ettiğimiz dataframe'i sıralayalım ve görelim bakallım Star Wars'a en yakın tavsiye edeceği film neymiş:
# Let's sort the dataframe we obtained and see which movie it recommends closest to Star Wars:
```

```python
corr_starwars.sort_values("Correlation", ascending=False).head(10)
```

# Görüldüğü gibi alakasız sonuyçlar çıktı, bu konuyu biraz araştırınca bunun nedeninin bu filmlerin çok az oy alması nedeniyle olduğunu bulacaksınız. Bu durumu düzeltmek için 100'den az oy alan filmleri eleyelim. Bu amaçla ratings isimli bir dataframe oluşturalım ve burada her filmin kaç tane oy aldığını (yani oy sayısını) tutalım.


# Görüldüğü gibi alakasız sonuyçlar çıktı, bu konuyu biraz araştırınca bunun nedeninin bu filmlerin çok az oy alması nedeniyle olduğunu bulacaksınız. Bu durumu düzeltmek için 100'den az oy alan filmleri eleyelim. Bu amaçla ratings isimli bir dataframe oluşturalım ve burada her filmin kaç tane oy aldığını (yani oy sayısını) tutalım.

```python
df.head()
```

```python
df.drop(["timestamp"], axis = 1)         # timestap sütununu kaldırdık.  ( # timestap sütununu kaldırdık.)
```

```python
# Her filmin ortalama (mean value) rating değerini bulalım. (# Let's find the mean value of each movie's rating.)

ratings = pd.DataFrame(df.groupby("title")["rating"].mean())   # Title'a göre grupluyor data DataFrame'in groupby özelliği ile. 

# Büyükten küçüğe sıralayıp bakalım (# Let's sort them from largest to smallest)

ratings.sort_values("rating", ascending=False).head()
```

```python
# Bu ortalamalar hesaplanırken kaç oy aldığına bakmadık o yüzden böyle hiç bilmediğimiz filmler çıktı.

# When calculating these averages, we did not look at how many votes they received, so we ended up with movies we knew nothing about.

ratings["rating_oy_sayisi"] = pd.DataFrame(df.groupby("title")["rating"].count())
ratings.head()
```

# Pandas'daki count metodu, genellikle bir veri çerçevesinde (DataFrame) veya bir seride (Series) eksik (NaN) değerler hariç, her bir sütunun veya serinin kaç tane geçerli (eksik olmayan) veri noktası içerdiğini saymak için kullanılır.


# # The count method in Pandas is typically used to count how many valid (non-missing) data points each column or series contains, excluding missing (NaN) values, in a dataframe or series.

```python
# Şimdi en çok oy alan filmleri büyükten küçüğre sıralayıp bakalım. 
# (Now let's take a look at the movies that received the most votes, in order from largest to smallest.)

ratings.sort_values("rating_oy_sayisi", ascending=False).head()
```

```python
# Tekrar esas amacımıza dönelim ve corr_starwars dataframe'imize göre rating_oy_sayisi sütununu ekleyelim. (join ile)
# Let's go back to our main purpose and add the rating_oy_sayisi column to our corr_starwars dataframe. (with join)
```

```python
corr_starwars.sort_values("Correlation", ascending=False).head(10)   # Bunla beraber gelen saçma sonuç veren dataframe'i düzelteceğiz.
# We will fix the dataframe that came with this that gave nonsensical results.
```

```python
corr_starwars = corr_starwars.join(ratings["rating_oy_sayisi"])
corr_starwars.head()
```
